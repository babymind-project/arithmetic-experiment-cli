```python
import data_utils
import matplotlib.pyplot as plt
from plot_results import *
```

```python
operator = 'add'
df_results = list()
result_files = get_all_result_files()
cols = ['data_index', 'answer', 'carries', 'solving_time']
for f in result_files:
    if get_operator(f) == operator:

        df_result = read_result_file(f)
        df_result = filter_for_mean_solving_time(df_result, rm_carry_outlier=False, outlier_std=2)
        df_result = df_result[df_result['correct'] == True]
        #df_result = df_result[df_result['solving_time'] < 30]
        df_result = df_result[cols]
        df_results.append(df_result)
        
        st_means = df_result.groupby(['carries'])['solving_time'].mean()
        st_stds = df_result.groupby(['carries'])['solving_time'].std()
        print_all = True
        if print_all:
            df_result.boxplot(column=['solving_time'], by='carries')
            plt.show()
            print(f)
            print('Means')
            print(st_means)
            print('STDs')
            print(st_means + 2 * st_stds)
            print(df_result[df_result['carries'] == 0].sort_values(by=['solving_time']))
            print(df_result[df_result['carries'] == 1].sort_values(by=['solving_time']))
            print(df_result[df_result['carries'] == 2].sort_values(by=['solving_time']))
            print(df_result[df_result['carries'] == 3].sort_values(by=['solving_time']))
            print(df_result[df_result['carries'] == 4].sort_values(by=['solving_time']))
        if st_means[2] > st_means[3] and (not print_all):
            print(f)
            print(st_means)
            print(st_means + 2 * st_stds)
            print(df_result[df_result['carries'] == 2].sort_values(by=['solving_time']))
            print(df_result[df_result['carries'] == 3].sort_values(by=['solving_time']))
            pass
```
